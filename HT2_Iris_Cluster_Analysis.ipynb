## Carga de datos y exploración inicial

Cargamos el dataset de Iris y verificamos sus primeras filas para familiarizarnos con la estructura de los datos. Los datasets utilizados son:

- **iris.csv**: Contiene las características de las flores Iris (longitud y ancho de los sépalos y pétalos).
- **iris-con-respuestas.csv**: Incluye las especies reales para la comparación posterior.

Además, se importan las librerías necesarias para el análisis.


```python
# Importar librerías necesarias para el análisis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from kneed import KneeLocator

# Cargar los datos desde los archivos CSV
df_iris = pd.read_csv("iris.csv")  # Dataset principal con características de las flores
df_iris_respuestas = pd.read_csv("iris-con-respuestas.csv")  # Dataset con las respuestas reales

# Mostrar las primeras filas del dataset principal para verificar la carga de datos
df_iris.head()
```

## Visualización inicial de los datos

Realizamos una **visualización exploratoria** del dataset usando gráficos de pares. Que nos permite identificar patrones iniciales, correlaciones entre las variables, y posibles agrupaciones de datos según las características de las flores (sépalos y pétalos).

- Se utiliza un gráfico de densidad en la diagonal para observar la distribución de cada característica.
- En las gráficas cruzadas, podemos observar las relaciones entre pares de variables.

Esta exploración ayudará a entender la estructura del dataset antes de aplicar técnicas de clustering.


```python
# Crear un gráfico de pares para visualizar las relaciones entre variables
sns.pairplot(df_iris, diag_kind='kde')  # diag_kind='kde' genera gráficos de densidad en la diagonal

# Agregar un título para la visualización
plt.suptitle("Visualización de los Datos - Iris Dataset", y=1.02)  # Ajustar la posición del título

# Mostrar la visualización
plt.show()
```

## Clustering con K-Means (2 Clusters)

Aplicamos el algoritmo **K-Means Clustering** para agrupar las observaciones del dataset en **2 clusters**. Este análisis nos ayudará a observar cómo K-Means separa los datos basándose en las características de los sépalos.

Pasos realizados:
1. Ajustamos el modelo de **K-Means** con 2 clusters.
2. Modificamos los nombres de las columnas para asegurar accesibilidad.
3. Visualizamos los resultados en un gráfico de dispersión, utilizando las características **longitud** y **ancho del sépalo**.


```python
# Aplicar el algoritmo de K-Means con 2 clusters
kmeans_2 = KMeans(n_clusters=2, random_state=42)  # Configuramos 2 clusters
df_iris['cluster_2'] = kmeans_2.fit_predict(df_iris.iloc[:, :-1])  # Asignamos cada punto a un cluster

# Asegurar que los nombres de columnas sean consistentes
df_iris.columns = df_iris.columns.str.strip().str.lower().str.replace(" ", "_")

# Graficar los resultados del clustering en función de los sépalos
sns.scatterplot(
    x=df_iris["sepal_length"],  # Eje X: longitud del sépalo
    y=df_iris["sepal_width"],   # Eje Y: ancho del sépalo
    hue=df_iris["cluster_2"],   # Colorear puntos según el cluster asignado
    palette="viridis"           # Paleta de colores
)

# Agregar título al gráfico
plt.title("Clustering con K-Means (2 Clusters)")

# Mostrar gráfico
plt.show()
```

## Estandarización de datos y clustering con K-Means (2 Clusters)

Aplicamos la **normalización de los datos** utilizando `StandardScaler` para garantizar que todas las variables tengan la misma escala. Este paso es crucial para algoritmos como K-Means, que son sensibles a las diferencias de escala entre las variables.

Pasos realizados:
1. Normalizamos las variables del dataset (longitud y ancho de sépalos y pétalos).
2. Volvemos a aplicar **K-Means Clustering** con 2 clusters.
3. Visualizamos los resultados con los datos normalizados para observar posibles diferencias respecto al análisis previo.


```python
# Normalización de los datos para garantizar que todas las variables tengan la misma escala
scaler = StandardScaler()  # Inicializamos el escalador estándar
df_iris_scaled = scaler.fit_transform(df_iris.iloc[:, :-1])  # Ajustamos y transformamos las características del dataset

# Aplicar K-Means nuevamente sobre los datos normalizados
kmeans_scaled = KMeans(n_clusters=2, random_state=42)  # Configuramos 2 clusters como antes
df_iris["cluster_2_scaled"] = kmeans_scaled.fit_predict(df_iris_scaled)  # Asignamos los clusters a los datos escalados

# Visualizar el clustering con los datos normalizados
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df_iris_scaled[:, 0],  # Eje X: primera característica escalada
    y=df_iris_scaled[:, 1],  # Eje Y: segunda característica escalada
    hue=df_iris["cluster_2_scaled"],  # Colorear según el cluster asignado
    palette="coolwarm"  # Usar una paleta de colores distinta para diferenciar
)

# Título del gráfico
plt.title("Clustering con Datos Normalizados (2 Clusters)")

# Mostrar gráfico
plt.show()
```

## Método del Codo para determinar el número óptimo de clusters

Utilizamos el **método del codo** para determinar el número óptimo de clusters. Este método evalúa cómo disminuye la inercia (distorsión) a medida que aumentamos el número de clusters.

Pasos realizados:
1. Calculamos la inercia (suma de distancias al centroide) para diferentes valores de `k` (de 1 a 10 clusters).
2. Graficamos la inercia contra el número de clusters.
3. Identificamos el "punto de inflexión" o "codo" en la gráfica, que sugiere el número óptimo de clusters.


```python
# Inicializamos una lista para almacenar los valores de inercia
inertia = []

# Iteramos sobre diferentes números de clusters (k)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)  # Configuramos K-Means con k clusters
    kmeans.fit(df_iris_scaled)  # Ajustamos el modelo a los datos normalizados
    inertia.append(kmeans.inertia_)  # Guardamos la inercia (distorsión) del modelo

# Graficamos el método del codo
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')  # Gráfica de inercia vs. número de clusters
plt.xlabel("Número de Clusters")  # Etiqueta del eje X
plt.ylabel("Inercia (Distorsión)")  # Etiqueta del eje Y
plt.title("Método del Codo para K-Means")  # Título del gráfico
plt.show()  # Mostramos la gráfica
```

## Comparación con datos reales

Comparamos el resultado del clustering obtenido mediante K-Means con las **especies reales** presentes en el dataset `iris-con-respuestas.csv`. Esto ayuda a evaluar la calidad del clustering y a determinar si los grupos generados por el modelo corresponden a las categorías reales de las especies.

Pasos realizados:
1. Se asignan códigos numéricos a las especies reales para facilitar la comparación.
2. Se genera un gráfico de dispersión con las etiquetas reales como referencia.
3. Se visualiza la separación de las observaciones según las especies.


```python
# Asignar códigos numéricos a las especies reales del dataset de respuestas
df_iris_respuestas["Cluster Real"] = df_iris_respuestas["species"].astype("category").cat.codes

# Graficar los datos reales para comparar con el clustering
sns.scatterplot(
    x=df_iris_scaled[:, 0],  # Eje X: primera característica escalada
    y=df_iris_scaled[:, 1],  # Eje Y: segunda característica escalada
    hue=df_iris_respuestas["Cluster Real"],  # Colorear según las especies reales
    palette="Dark2"  # Paleta de colores
)

# Título del gráfico
plt.title("Comparación con Datos Reales")

# Mostrar gráfico
plt.show()
```

## Clustering usando solo la forma del pétalo

Aplicamos **K-Means Clustering** utilizando únicamente las características del **pétalo**: longitud y ancho. Esto nos permite observar cómo estas variables específicas separan las observaciones y compararlo con el clustering basado en las características del sépalo.

Pasos realizados:
1. Seleccionamos solo las variables `petal_length` y `petal_width`.
2. Normalizamos las variables seleccionadas para garantizar escalas consistentes.
3. Aplicamos **K-Means Clustering** con 3 clusters.
4. Visualizamos los resultados en un gráfico de dispersión.


```python
# Seleccionar las características del pétalo para el análisis
df_petal = df_iris[['petal_length', 'petal_width']]  # Usar nombres corregidos

# Normalizar las características seleccionadas
df_petal_scaled = scaler.fit_transform(df_petal)  # Escalado para mantener consistencia

# Aplicar K-Means con 3 clusters a las características del pétalo
kmeans_petal = KMeans(n_clusters=3, random_state=42).fit(df_petal_scaled)
df_iris['cluster_petal'] = kmeans_petal.labels_  # Guardar los clusters en el DataFrame

# Graficar los resultados del clustering basado en el pétalo
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df_petal_scaled[:, 0],  # Eje X: longitud del pétalo escalada
    y=df_petal_scaled[:, 1],  # Eje Y: ancho del pétalo escalado
    hue=df_iris["cluster_petal"],  # Colorear según el cluster asignado
    palette="coolwarm"  # Paleta de colores
)

# Título del gráfico
plt.title("Clustering usando solo Forma del Pétalo")

# Mostrar gráfico
plt.show()
```

## Determinación del número óptimo de clusters con `kneed`

Utilizamos la librería **`kneed`** para encontrar el "punto de inflexión" o "codo" en la gráfica de inercia calculada anteriormente. Este método identifica automáticamente el número óptimo de clusters basado en el comportamiento de la curva.

Pasos realizados:
1. Aplicamos el objeto `KneeLocator` para analizar los valores de inercia calculados para diferentes valores de `k`.
2. Usamos el parámetro `curve="convex"` y `direction="decreasing"` para ajustar el análisis.
3. Mostramos el número óptimo de clusters según la librería `kneed`.


```python
# Usar KneeLocator para determinar el número óptimo de clusters
kl = KneeLocator(range(1, 11), inertia, curve="convex", direction="decreasing")

# Imprimir el número óptimo de clusters identificado por kneed
print("Número óptimo de clusters según kneed:", kl.elbow)
```
